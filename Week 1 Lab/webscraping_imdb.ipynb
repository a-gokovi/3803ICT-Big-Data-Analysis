```python
# Write a Python script to download IMDB 250 Top Rated Movies.
# For each movie, you'll need to retrieve the movie title, the ranking, the initial release year, the casting and 
# the rating.
# Your data must be stored in a proper imdb_top_250.csv file.

import requests, csv
from bs4 import BeautifulSoup

def get_movies(url):
    response = requests.get(url)
    soup_imdb = BeautifulSoup(response.text)

    all_movies = soup_imdb.find_all("td", {'class': 'titleColumn'})

    all_ratings = soup_imdb.find_all("strong")
   
   #initialise lists for each information
    titles = []
    castings = []
    years = []
    rankings = [x for x in range(1, len(all_movies) + 1)]
    ratings =[]


    for index in range(len(all_movies)):
        #Get casting
        castings.append(all_movies[index].a['title'])
        #Get years, removing brackets
        years.append(all_movies[index].find('span', class_ = "secondaryInfo").text.strip("()"))
        #Get title
        titles.append(all_movies[index].find('a').text.strip())
        #Get rating
        ratings.append(float(all_ratings[index].text))

    mega_list = list(zip(titles, rankings, years, castings, ratings))
    mega_list = [list(elt) for elt in mega_list]

    # Write csv
    str_lst = [f"{mega_list[i][1]} / {mega_list[i][0]} ({mega_list[i][2]}) / Starring: {mega_list[i][3]}" \
    for i, val in enumerate(mega_list)]

    res = [elt.split("/") for elt in str_lst]

    with open("imdb_top_250.csv", "w", encoding='utf-8') as f:
        writer = csv.writer(f, delimiter="-")
        writer.writerows(res)
```

```python
def main():
    url = "https://www.imdb.com/chart/top"
    get_movies(url)
    

if __name__ == "__main__":
    main()
    
```
